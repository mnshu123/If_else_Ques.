```python
#Q1. Calculate the electricity bill ?

units = int(input("Enter number of units"))
if units<0:
    print("incorrect data")
bill =0
if units>200:
    bill= (units -200)*10
elif units > 100:
    bill = (units - 100)*5
print("Total bill:Rs",bill)  
```

```python
#Q2. Program to accept percentage from the user and display the grade according to the folloing criteria ? 

percentage = float(input("enter your percentage"))
if percentage<0 or percentage >100:
    print("not found")
elif percentage>90:
    print("Grade:A")
elif percentage>80:
    print("Grade:B")
elif percentage>=60:
    print("Grade:C")
else:
    print("Grade:D")
```

```python
#Q3.Accept the age of 4 people and display the youngest one ?

age1=int(input("Enter your age: "))
age2=int(input("Enter your age: "))
age3=int(input("Enter your age: "))
ages=[age1,age2,age3]
youngest = max(ages)
print("the youngest age is",youngest) 
```

```python
#Q5. Second largest number ?

a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
c=int(input("Enter third number:"))
nums=[a,b,c]
nums.sort()
print("second largest number",nums[1])
```

```python
#Q4. Decided to give bonus to employee according to following criteria ?

salary = float(input("Enter your salary:"))
year = float(input("Enter your Time of Service:"))
if year > 10:
    bonus = 10/100*salary
elif year >= 6:
    bonus = 8/100*salary
elif year < 6:
    bonus = 5/100*salary
print("Net_bonus",bonus)
print("Total_bonus",salary+bonus)
```

```python
#Q5. Accept three numbers from user and display the second largest number ?

a = float(input("Enter your 1st number:"))
b = float(input("Enter your 2st number:"))
c = float(input("Enter your 3st number:"))

num=[a,b,c]
num.sort()
print("The second largest number",num[1])
```

```python
#Q6. Calculate Net_Amount ?

Market_price = float(input("Enter your price:"))
if Market_price > 10000:
    Discount = 20/100*Market_price 
elif Market_price > 7000:
    Discount =  15/100*Market_price 
elif Market_price < 7000:
    Discount = 10/100 *Market_price 
Net_amount = Market_price - Discount
print("Discount:",Discount)
print("Net Amount to pay",Net_amount)
    
```
